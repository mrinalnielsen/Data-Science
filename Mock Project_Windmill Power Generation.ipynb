```python
#Displaying outliers present in 'wind_speed(m/s)'
outliers = []
def detect_outliers_iqr(PowerData):
    PowerData = sorted(PowerData)
    q1 = np.percentile(PowerData, 25)
    q3 = np.percentile(PowerData, 75)
    # print(q1, q3)
    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    # print(lwr_bound, upr_bound)
    for i in PowerData: 
        if (i<lwr_bound or i>upr_bound):
            outliers.append(i)
    return outliers# Driver code
sample_outliers_ws = detect_outliers_iqr(PowerData['wind_speed(m/s)'])
print("Outliers from IQR method: ", sample_outliers_ws)
```

```python
#Displaying percentage of outliers present in 'wind_speed(m/s)'
print ('percentage_outliers:',len(sample_outliers_ws )/len(PowerData))
```

```python
#Imputing the outliers of each column with their respective upper bound and lower bound

for i in PowerData.columns:
    if (PowerData[i].dtype!='object'):
        Q1 = PowerData[i].quantile(0.25)
        Q3 = PowerData[i].quantile(0.75)
        iqr = Q3-Q1
        upper_limit = Q3 + 1.5 * iqr
        lower_limit = Q1 - 1.5 * iqr
        PowerData[i] = np.where(PowerData[i] >= upper_limit,upper_limit,np.where(PowerData[i] <= lower_limit,lower_limit,PowerData[i]))
PowerData
```
